```python
import pandas as pd 
from collections import Counter
df = pd.read_excel (r'C:\Users\Anar\Desktop\Code Academy\Week 8\Turbo_az_Kia_all_vehicles.xlsx')
print (df)
```

```python
# Drop duplicate rows based on all columns
df_no_duplicates = df.drop_duplicates()

# Display the DataFrame without duplicates
print(df_no_duplicates)
```

```python
 #! Most Frequent 1 - 10 Less Frequent

# Count occurrences of each value in city
city_frequencies = df['City'].value_counts()

# Convert the result to a dictionary
city_frequencies_dict = city_frequencies.to_dict()

# Sort the cities by their frequency in descending order
sorted_cities = sorted(city_frequencies_dict.items(), key=lambda x: x[1], reverse=True)

# Assign ranks to the cities
ranked_cities = {}
for i, (city, freq) in enumerate(sorted_cities):
    ranked_cities[city] = i + 1 if i < 10 else 0

# Apply the ranking to the 'City' column
df['City'] = df['City'].map(ranked_cities)

```

```python
#Drop the Brand column
df.drop(columns = ['Brand', 'Product_link', 'Saler_name'], inplace = True)
print(df)
```

```python
# Count occurrences of each value in the 'Model' column and normalize the frequencies
model_frequencies = df['Model'].value_counts(normalize=True).to_dict()

# Round the percentages to two decimal places
rounded_model_frequencies = {model: round(frequency, 2) for model, frequency in model_frequencies.items()}

# Map the rounded frequencies to the 'Model' column
df['Model'] = df['Model'].map(rounded_model_frequencies)
```

```python
# Count occurrences of each value in year
year_frequencies = df['Year'].value_counts(normalize=True).to_dict()

# Round the percentages to two decimal places
rounded_year_frequencies = {year: round(frequency, 2) for year, frequency in year_frequencies.items()}

# Map the year frequencies to the 'Year' column
df['Year'] = df['Year'].map(rounded_year_frequencies)
```

```python
# Count occurrences of each value in city Body Type
body_type_frequencies = df['Body Type'].value_counts(normalize=True).to_dict()

# Round the percentages to two decimal places
rounded_body_type_frequencies = {bodytype: round(frequency, 2) for bodytype, frequency in body_type_frequencies.items()}

# Map the body type frequencies to the 'Body Type' column
df['Body Type'] = df['Body Type'].map(rounded_body_type_frequencies)
```

```python
# Count occurrences of each value in city Color
color_frequencies = df['Color'].value_counts(normalize=True).to_dict()

# Round the percentages to two decimal places
rounded_color_frequencies = {color: round(frequency, 2) for color, frequency in color_frequencies.items()}

# Map the color frequencies to the 'Color' column
df['Color'] = df['Color'].map(rounded_color_frequencies)
```

```python
split_engine = df['Engine Details'].str.split('/',expand = True)
split_engine
df['engine_capacity'] = split_engine[0]
df['horse_power'] = split_engine[1]
df['fuel_type'] = split_engine[2]
df.drop(columns=['Engine Details'], inplace=True)
df
```

```python
df['engine_capacity'] = df['engine_capacity'].str.replace(' L','')
df['engine_capacity'] = pd.to_numeric(df['engine_capacity'])
df
```

```python
df['horse_power'] = df['horse_power'].str.replace(' a.g.','')
df['horse_power'] = pd.to_numeric(df['horse_power'])
df
```

```python
# Count occurrences of each value in city Fuel Type
fuel_type_frequencies = df['fuel_type'].value_counts(normalize=True).to_dict()

# Round the percentages to two decimal places
rounded_fuel_type_frequencies = {fueltype: round(frequency, 2) for fueltype, frequency in fuel_type_frequencies.items()}

# Map the fuel_type frequencies to the 'fuel_type' column
df['fuel_type'] = df['fuel_type'].map(rounded_fuel_type_frequencies)
```

```python
# Clean the 'Mileage' column by removing non-numeric characters and converting to numeric
df['Mileage'] = df['Mileage'].str.replace(r'\D', '', regex=True)
df['Mileage'] = pd.to_numeric(df['Mileage'])

# Display the DataFrame after cleaning and conversion
print(df)
```

```python
# Count occurrences of each value in city Transmission
Transmission_frequencies = df['Transmission'].value_counts(normalize=True).to_dict()

# Round the percentages to two decimal places
rounded_Transmission_frequencies = {Transmission: round(frequency, 2) for Transmission, frequency in Transmission_frequencies.items()}

# Map the Transmission frequencies to the 'Transmission' column
df['Transmission'] = df['Transmission'].map(rounded_Transmission_frequencies)
```

```python
# Count occurrences of each value in city Drive Type

drive_type_frequencies = df['Drive Type'].value_counts(normalize=True).to_dict()

# Round the percentages to two decimal places
rounded_drive_type_frequencies = {drive_type: round(frequency, 2) for drive_type, frequency in drive_type_frequencies.items()}

# Map the Drive Type frequencies to the 'Drive Type' column
df['Drive Type'] = df['Drive Type'].map(rounded_drive_type_frequencies)
```

```python
# Print unique values in the 'Is_New' column
unique_values = df['Is_New'].unique()
print(unique_values)

# Replace specific values with dummy values 
dummy_values = {'Xeyr': 0, 'Bəli': 1,}

df['Is_New'] = df['Is_New'].replace(dummy_values)

print(df)
```

```python
# Replace 'deger yok' with 5 in the 'seat_count' column
df['Seat Count'] = df['Seat Count'].replace('değer yok', 5)
df['Seat Count'] = df['Seat Count'].replace('8+', 8)
# Display the DataFrame after replacing values
print(df)
```

```python
#Drop the Owner column
df.drop(columns = ['Owner Count'], inplace = True)
print(df)
```

```python
# Filter out rows where 'ownership' column is not equal to 'deger yok'
df = df[df['Condition'] != 'değer yok']

# Display the DataFrame after filtering
print(df)
```

```python
split_condition = df['Condition'].str.split(',', n=1, expand=True)

# Use .loc to set values in the original DataFrame
df.loc[:, 'accidental'] = split_condition[0]
df.loc[:, 'colored'] = split_condition[1]
df.drop(columns=['Condition'], inplace=True)
df
```

```python
# Print unique values in the 'colored' column
unique_values_color = df['colored'].unique()
print(unique_values_color)

# Define dummy values mapping
dummy_values_color = {' rənglənib': 0, ' rənglənməyib': 1}

# Replace specific values with dummy values 
df['colored'] = df['colored'].replace(dummy_values_color)

df
```

```python
# Print unique values in the 'accidental' column
unique_values = df['accidental'].unique()
print(unique_values)

# Replace specific values with dummy values 
dummy_values = {'Vuruğu var': 0, 'Vuruğu yoxdur': 1,}

df['accidental'] = df['accidental'].replace(dummy_values)

df
```

```python
# Step 2: Count the frequency of each Origin
origin_counts = df['Origin'].value_counts()

# Step 3: Map each origin to a number from 0 to 50 based on its frequency
origin_mapping = {origin: min(idx, 50) for idx, (origin, _) in enumerate(origin_counts.items(), start=1)}
origin_mapping['değer yok'] = 0

# Apply the mapping to the 'Origin' column
df['Origin'] = df['Origin'].map(origin_mapping)

df
```

```python
# Split 'Price' into 'amount' and 'currency'
split_price = df['Price'].str.extract(r'(\d+(?:\s\d+)?)\s+(AZN|USD)')

# Rename columns for clarity
split_price.columns = ['amount', 'currency']

# Convert 'amount' to numeric
split_price['amount'] = split_price['amount'].str.replace(' ', '').astype(float)

# Define exchange rate for USD to AZN conversion
exchange_rate_usd_to_azn = 1.7

# Convert USD prices to AZN
usd_mask = split_price['currency'] == 'USD'
split_price.loc[usd_mask, 'amount'] *= exchange_rate_usd_to_azn
split_price.loc[usd_mask, 'currency'] = 'AZN'

# Drop the 'currency' column as it's no longer needed
split_price.drop(columns=['currency'], inplace=True)

# Rename the 'amount' column to 'new_price'
split_price.rename(columns={'amount': 'new_price'}, inplace=True)

# Concatenate the split_price DataFrame back to the original DataFrame
df = pd.concat([df, split_price], axis=1)

# Drop the original 'Price' column
df.drop(columns=['Price'], inplace=True)

df
```

```python
# Convert all columns to numeric data type
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Extract the correlation of 'Price' with other columns
price_correlation = correlation_matrix['new_price']

# Print the correlation of 'Price' with other columns
print("Correlation of 'new_price' with other columns:")
print(price_correlation)
```
