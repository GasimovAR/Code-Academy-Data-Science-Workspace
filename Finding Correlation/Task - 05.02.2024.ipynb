```python
import pandas as pd 

df = pd.read_excel(r'C:\Users\Anar\Desktop\test_dataset_new.xlsx')

print(df)
```

```python
# Count 'Not found' occurrences in each column
not_found_counts = df.eq('Not found').sum()

print(not_found_counts)
```

```python
#Delete the not found from the price column
def delete_rows_with_not_found(df, column_name, price_to_delete):
    new_price = df[column_name] != price_to_delete
    df = df[new_price]
    
    return df
# Specify the prices you want to delete
price_to_delete = 'Not found'

df = delete_rows_with_not_found(df, 'price', price_to_delete)

print(df)
```

```python
# Function to split price column into currency and amount
def split_price_column(df):
    # Remove the first dot and keep the second dot
    df['amount'] = df['price'].str.replace('.', '', 1)
    # Replace commas with dots for all values
    df['amount'] = df['price'].str.replace(',', '.', regex=False)
    # Splitting the 'price' column into 'amount' and 'currency' columns
    df[['amount', 'currency']] = df['price'].str.split(' ', expand=True)
    return df

# Call the function to split the price column
df = split_price_column(df)

print(df)
```

```python
# Delete the 'currency' column
df.drop(columns=['currency'], inplace=True)

print(df)
```

```python
# Convert 'amount' column from string to numeric
df['amount'] = pd.to_numeric(df['amount'], errors = 'coerce')

print(df)
```

```python
# Function to replace 'Not found' with 0
def not_found_to_zero(df):
    df['favorite-count'] = df['favorite-count'].str.replace('Not found', '0')

# Call the function
not_found_to_zero(df)

# Print the DataFrame after modification
print(df)
```

```python
# Function to add 50 to price if it's >= 150
def add_cargo(amount):
    if amount <= 150:
        return amount + 50
    else:
        return amount

# Apply the function to the 'price' column and create a new column 'price_cargo'
df['price_cargo'] = df['amount'].apply(add_cargo)

print(df)
```

```python
# Print unique values in the 'dd-txt-vl' column
unique_values = df['dd-txt-vl'].unique()
print(unique_values)

# Replace specific values with dummy values 
#! Good 7-6-5-4-3-2-1-0 Bad
dummy_values = {'Not found': 0, '2 gün içinde': 6, '1 gün içinde kargoda': 7, 
                '5 gün içinde': 3, '10 gün içinde': 2, '3 gün içinde': 5, 
                '4 gün içinde': 4, '30 gün içinde': 1}

df['dd-txt-vl'] = df['dd-txt-vl'].replace(dummy_values)

print(df)
```

```python
# Calculate the average of the other values in the 'sl-pn' column
filtered_ratings = df['sl-pn'][df['sl-pn'] != 'Not found'].astype(float)
average_rating = filtered_ratings.mean()

# Replace 'Not found' with the calculated average
df['sl-pn'] = df['sl-pn'].replace('Not found', average_rating)

print(df)
```

```python
# Step 1: Replace empty cells with 'Unknown'
df['Renk'].fillna('Unknown', inplace=True)

# Step 2: Count the frequency of each Renk
color_counts = df['Renk'].value_counts()

# Step 3: Map each Renk to a number from 0 to 50 based on its frequency
color_mapping = {color: min(idx, 50) for idx, (color, _) in enumerate(color_counts.items(), start=1)}
color_mapping['Unknown'] = 0

# Apply the mapping to the 'Renk' column
df['Renk'] = df['Renk'].map(color_mapping)

print(df)
```
